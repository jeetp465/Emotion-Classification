```python
import requests
import os
```

```python
def download_images(filename, target_dir):
    with open(filename) as f:
        url = f.readlines()

    url = [elem.strip('\n') for elem in url]
    url = [[elem.split(' ')[1], elem.split(' ')[0]] for elem in url]

    url_dict = {}
    for key, value in url:
        url_dict.setdefault(key, []).append(value)
        
    for label, urls in url_dict.items():
        for url in urls:
            filename = url.split('_',2)[-1]
            pathname = target_dir + '/' + str(label) + '/' + filename

            if os.path.isfile(pathname):
                continue
            else:
                response = requests.get(url)
                if response.status_code == 200:
                    with open(pathname, 'ab') as f:
                        f.write(response.content)
```
